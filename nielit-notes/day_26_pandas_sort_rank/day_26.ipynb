```python
# 0 import pandas 

import pandas as pd
import numpy as np
```

```python
# 1 create a dataframe object from a (10,4) 2D array
# consisting of random integers between 10 and 20 by making c1,c2,c3,c4 as column indices.

arr = np.random.randint(10, 20, 40).reshape(10, 4)
df = pd.DataFrame(arr, columns='c1 c2 c3 c4'.split())
df
```

```python
# 2 Sort the above oject based on the key 'c1' in ascending and then by 'c3' in descending order.

df.sort_values(['c1', 'c3'])
```

```python
#  3. Write script to store item, place and total sale in a dataframe object.
#     (i)   There should be 3 or more places and 4 or more items in the set. Based on the choice entered by user
#     (ii)  Show placewise rank for a particular item (for a given item name).
#     (iii) Show itemwise rank for a particular place (for a given place name).

def data(item, place, total):
    return pd.DataFrame({'item' : item, 'place' : place, 'total' : total})

def place_item_rank(df, item):
    df = df[df['item']==item]
    df = df.sort_values(['total'], ascending=False)
    df['rank'] = df['total'].rank()
    return df[['item', 'place', 'total', 'rank']]

def item_place_rank(df, place):
    df = df[df['place']==place]
    df = df.sort_values(['total'], ascending=False)
    df['rank'] = df['total'].rank()
    return df[['place', 'item', 'total', 'rank']]

item = ['pen', 'pencil', 'book', 'scale', 
        'pen', 'pencil', 'book', 'scale',
        'pen', 'pencil', 'book', 'scale',
        'pen', 'pencil', 'book', 'scale']

place = ['tvm', 'tvm', 'tvm', 'tvm', 
         'enk', 'enk', 'enk', 'enk',
         'cal', 'cal', 'cal', 'cal', 
         'knr', 'knr', 'knr', 'knr']

total = np.random.randint(10, 50, 16)
```

```python
df = data(item, place, total)
df.head()
```

```python
pi_rank = place_item_rank(df, 'pen')
pi_rank.head()
```

```python
ip_rank = item_place_rank(df, 'tvm')
ip_rank.head()
```
