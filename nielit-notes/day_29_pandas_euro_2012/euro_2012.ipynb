```python
# 1. Import the necessary libraries(pandas)

import pandas as pd
```

```python
# 2. Import the dataset euro2012 from common folder.
# 3. Assign it to a data frame variable called euro12.

df = pd.read_csv('euro2012.csv')
df.head()
```

```python
# 4. Select only the Goal column.

df['Goals']
```

```python
#  5. How many team participated in the Euro2012?

# len(df)
# df.shape
df['Team'].nunique()
```

```python
# 6. What is the number of columns in the dataset?

# df.shape[1]
len(df.columns)
```

```python
#  7. View only the columns Team, Yellow Cards and Red Cards 
#     and assign them to a dataframe called discipline

discipline = df[['Team', 'Yellow Cards', 'Red Cards']]
discipline.head()
```

```python
#  8. Sort the teams by Red Cards, then to Yellow Cards

discipline.sort_values(['Red Cards', 'Yellow Cards'], ascending=False).reset_index(drop=True)[['Team', 'Red Cards', 'Yellow Cards']].head()
```

```python
#  9. Calculate the mean Yellow Cards given per Team

df.groupby('Team')['Yellow Cards'].mean()
```

```python
#  10. Filter teams that scored more than 6 goals

df[df['Goals']>6]
```

```python
# 11. Select the teams that start with G

df[df['Team'].str.startswith('G')]
```

```python
# 12. Select the first 7 columns

df.iloc[:, 0:7].head()
```

```python
# 13. Select all columns except the last 3.

df.iloc[:, :-3].head()
```

```python
#  14. Present only the Shooting Accuracy from England, Italy and Russia

# df[['Team', 'Shooting Accuracy']].query("Team == 'England' or Team == 'Italy' or Team == 'Russia'")
df[df['Team'].isin(['England', 'Italy', 'Russia'])][['Team', 'Shooting Accuracy']]
```
